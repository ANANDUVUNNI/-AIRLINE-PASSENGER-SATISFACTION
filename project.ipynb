```python
import pandas as pd
```

```python
df=pd.read_csv("/content/airline_passenger_satisfaction.csv")
df
```

```python
df.isna().sum()

```

```python
df["arrival_delay_in_minutes"]=df["arrival_delay_in_minutes"].fillna(df["arrival_delay_in_minutes"].mean())
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["Gender"]=le.fit_transform(df["Gender"])
df["customer_type"]=le.fit_transform(df["customer_type"])
df["type_of_travel"]=le.fit_transform(df["type_of_travel"])
df["customer_class"]=le.fit_transform(df["customer_class"])
df["arrival_delay_in_minutes"]=le.fit_transform(df["arrival_delay_in_minutes"])
df["satisfaction"]=le.fit_transform(df["satisfaction"])


```

```python
df.dtypes
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python

#oversampling
over=SMOTE()
x,y=over.fit_resample(x,y)
```

```python
y.value_counts()
```

```python
scaler=MinMaxScaler()
scaler.fit(x)
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.3,random_state=1)
```

```python
#apply algorithms
knn=KNeighborsClassifier()
rf=RandomForestClassifier()
ab=AdaBoostClassifier()
sv=SVC()
lst=[knn,rf,ab,sv]
for i in lst:
  print(i)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print(classification_report(y_test,y_pred))
```

```python
#tune random forest using GridSearchCV
rt1=RandomForestClassifier()
parameter={"criterion":["gini","entropy","log_loss"],
           "max_features":["sqrt","log2"],
           "class_weight":["balanced","balanced_subsample"]}
clf=GridSearchCV(rt1,parameter,cv=10,scoring="accuracy")
clf.fit(x_train,y_train)
```

```python
print(clf.best_params_)
```

```python
rf=RandomForestClassifier(n_estimators=100,criterion="gini",max_features="log2")
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
print(classification_report(y_test,y_pred))
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
